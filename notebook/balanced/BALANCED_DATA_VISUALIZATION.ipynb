```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
merge_df=pd.read_csv('/content/merge_datasets.csv')
balanced_df=pd.read_csv('/content/allbalanced_dataset.csv')
```

BAR CHART CLEANED DATASET

```python
x=merge_df['rating'].value_counts().index
y=merge_df['rating'].value_counts().values
sns.barplot(x=x,y=y)
plt.title('bar plot-ratings count')
plt.xlabel('ratings')
plt.ylabel('count')
plt.show()
```

PIE CHART

```python
merge_df['rating'].value_counts().sort_index().plot.pie(autopct='%1.1f%%',startangle=90,colors=sns.color_palette('bright'))
plt.title('pie chart-ratings count')
plt.ylabel('')
plt.show()
```

HORIZONTAL BAR

```python
plt.barh(x,y)
plt.title('horizontal bar plot')
plt.xlabel('count')
plt.ylabel('ratings')
plt.show()
```
