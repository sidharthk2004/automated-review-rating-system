```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "Review"] = df["Review"].apply(clean_text)
```

COLUMNS UPDATE TO INT

```python
df['Rating']=df['Rating'].astype(int)
```
