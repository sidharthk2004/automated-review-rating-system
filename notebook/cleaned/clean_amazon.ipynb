```python
df.drop(columns=['product_id', 'product_name', 'category', 'discounted_price',
       'actual_price', 'discount_percentage',  'rating_count',
       'about_product', 'user_id', 'user_name', 'review_id', 'review_title',
                 'img_link', 'product_link'],inplace=True)
```

```python
df.isnull().sum()#to identify how much null values are there
```

```python
df.dropna(inplace=True)
```

```python
df.rename(columns={'review_content':'review'},inplace=True)
```

```python
df.head(3)
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(subset=['review','rating'],inplace=True)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "review"] = df["review"].apply(clean_text)
```

COLUMNS UPDATE TO INT

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce').fillna(0).astype(int)
```

```python
df['rating'].value_counts()
```

```python
df=df[df['rating']!=0]
```
