```python
df.drop(columns=['Unnamed: 0', 'drug_name', 'date', 'age', 'gender', 'time_on_drug',
       'reviewer_type', 'condition',  'rating_effectiveness',
       'rating_ease_of_use', 'rating_satisfaction', ],inplace=True)
```

```python
df.isnull().sum() #to identify how much null values are there
```

```python
df.dropna(inplace=True) #to drop null values
```

```python
df.rename(columns={'rating_overall':'rating','text':'review'},inplace=True)
```

```python
df.head(2)
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(subset=['review','rating'],inplace=True)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "review"] = df["review"].apply(clean_text)
```

COLUMNS UPDATE TO INT

```python
df['rating']=df['rating'].astype(int)
```
