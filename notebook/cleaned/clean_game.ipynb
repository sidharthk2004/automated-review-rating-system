```python
df.drop(columns=['id', 'title', 'game_url', 'image_url', 'system', 'publisher',
       'developer', 'genre', 'num_players', 'release_date', 'review', 'num_votes'],inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.rename(columns={'review_conclusion':'REVIEW'},inplace=True)
```

```python
df.head(5)
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(subset=['REVIEW','rating'],inplace=True)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "REVIEW"] = df["REVIEW"].apply(clean_text)
```
