```python
df.drop(columns=['reviewerID', 'asin', 'reviewerName', 'helpful', 'reviewText',
        'unixReviewTime', 'reviewTime'],inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df.rename(columns={'overall':'rating','summary':'review'},inplace=True)
```

```python
df.head(2)
```

```python
print(df.duplicated().sum())
```

```python
df.drop_duplicates(subset=['review','rating'],inplace=True)
```

```python
import re

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r"http\S+", "", text)
    text = re.sub(r"[^\w\s]", "", text)
    text = re.sub(r"\d+", "", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text

df.loc[:, "review"] = df["review"].apply(clean_text)
```
